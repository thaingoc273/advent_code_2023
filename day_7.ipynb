```python
def classify_card(card):
    set_char = set(list(card))
    dict_char = {}
    for char in set_char:
        count_char = card.count(char)
        dict_char[char] = count_char
    if len(set_char) == 1:
       return 1 # Five of kind
    elif len(set_char) == 2:
        for char in set_char:
            if dict_char[char]==4:
                return 2 # Four of kind
        return 3 # Full house
    elif len(set_char) == 3:
        for char in set_char:
            if dict_char[char] == 3:
                return 4 # Three of a kind
        return 5 # Two pair
    elif len(set_char) == 4:
        return 6 # One pair
    else:
        return 7 # High card
```

```python
card_power = ['A', 'K', 'Q', 'J', 'T', '9', '8', '7', '6', '5', '4', '3', '2']
```

```python
def get_char_power(card):
    return tuple(card_power.index(c) for c in card)
```

```python
def get_point(lst_card):
    lst_card_rank = []
    dict_card = {card: number for card, number in lst_card}
    for card, _ in lst_card:
        lst_card_rank.append((card, classify_card(card), get_char_power(card)))

    lst_card_rank.sort(key=lambda x:(x[1], x[2]), reverse=True)

    winning_point = 0
    for i in range(len(lst_card_rank)):
        winning_point += (i+1)*dict_card[lst_card_rank[i][0]]
        
    return winning_point
```

```python
card_power_new = ['A', 'K', 'Q', 'T', '9', '8', '7', '6', '5', '4', '3', '2', 'J']
```

```python
def replace_jocker(card):
    n = card.count('J')
    if n==0:
        return  card
    elif card=='JJJJJ':
        return 'AAAAA'
    else:
        tupl_char_count = [(card.count(char), char) for char in card if char!='J']
        tupl_char_count = [(count, card_power_new.index(char)) for count, char in tupl_char_count]
        tupl_char_count.sort(key = lambda x: (-x[0], x[1]))
        change_char_index = tupl_char_count[0][1]
        
        return card.replace('J', card_power_new[change_char_index])
```

```python
from collections import Counter

with open("data/day_7.txt") as f:
    data = f.read().strip()


def hand_type(hand):
    c = Counter(hand)
    counts = [0] if (jokers := c.pop("*", 0)) == 5 else sorted(c.values())
    # The most efficient use of a joker is always as the most common non-joker card
    counts[-1] += jokers
    match counts:
        case *_, 5:
            return 7
        case *_, 4:
            return 6
        case *_, 2, 3:
            return 5
        case *_, 3:
            return 4
        case *_, 2, 2:
            return 3
        case *_, 2:
            return 2
    return 1


def solve(data):
    ws = [l.split() for l in data.split("\n")]
    return sum(
        rank * bid
        for rank, (*_, bid) in enumerate(
            sorted(
                (hand_type(hand), *map("*23456789TJQKA".index, hand), int(bid))
                for hand, bid in ws
            ),
            1,
        )
    )


# Part 1
print(solve(data))

# Part 2
print(solve(data.replace("J", "*")))
```
