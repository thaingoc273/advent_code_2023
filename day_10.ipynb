```python
lst_pipes = ['|', '-', 'L', 'J', '7', 'F']
# Write a function with previous_pipe = (x, y, pipe)

def find_next_pipe(previous_pipe, double_previous):
    if previous_pipe[2]=='-':
        if double_previous[1]>previous_pipe[1]:
            x_next = previous_pipe[0]
            y_next = previous_pipe[1]-1            
        else:
            x_next = previous_pipe[0]
            y_next = previous_pipe[1]+1     
    elif previous_pipe[2] == '|':
        if double_previous[0]>previous_pipe[0]:
            x_next = previous_pipe[0]-1
            y_next = previous_pipe[1] 
        else:
            x_next = previous_pipe[0]+1
            y_next = previous_pipe[1] 
    elif previous_pipe[2] == 'L':
        if double_previous[0] == previous_pipe[0]:
            x_next = previous_pipe[0]-1
            y_next = previous_pipe[1]
        else:
            x_next = previous_pipe[0]
            y_next = previous_pipe[1]+1
    elif previous_pipe[2] == 'J':
        if double_previous[0] == previous_pipe[0]:
            x_next = previous_pipe[0]-1
            y_next = previous_pipe[1]
        else:
            x_next = previous_pipe[0]
            y_next = previous_pipe[1]-1
    elif previous_pipe[2] == '7':
        if double_previous[0] == previous_pipe[0]:
            x_next = previous_pipe[0]+1
            y_next = previous_pipe[1]
        else:
            x_next = previous_pipe[0]
            y_next = previous_pipe[1]-1
    elif previous_pipe[2] == 'F':
        if double_previous[0] == previous_pipe[0]:
            x_next = previous_pipe[0]+1
            y_next = previous_pipe[1]
        else:
            x_next = previous_pipe[0]
            y_next = previous_pipe[1]+1
    return (x_next, y_next, arr_tiles[x_next, y_next])
```

```python
# Check the pipe 'S'
x_start = np.where(arr_tiles=='S')[0][0]
y_start = np.where(arr_tiles=='S')[1][0]
start = (x_start, y_start)
arr_tiles[x_start, y_start]
max_x = arr_tiles.shape[0]
max_y = arr_tiles.shape[1]
```

```python
# Finding all possible move from start
def check_ind(x, y):
    return (x>=0) and (x<max_x) and (y>=0) and (y<max_y)

def next_move(start):
    lst_move = []
    x = start[0]
    y = start[1]
    if check_ind(x+1, y) and arr_tiles[x+1, y]!='.':
        lst_move.append((x+1, y, 'S'))
    if check_ind(x-1, y) and arr_tiles[x-1, y]!='.':
        lst_move.append((x-1,y, 'N'))
    if check_ind(x, y+1) and arr_tiles[x, y+1]!='.':
        lst_move.append((x, y+1, 'E'))
    if check_ind(x, y-1) and arr_tiles[x, y-1]!='.':
        lst_move.append((x, y-1, 'W'))
    return lst_move    
```

```python
dict_tile ={'N': '|7F', 'S': '|LJ', 'E': '-7J', 'W': '-LF'}
```

```python
def finding_start(start):
    lst_move = next_move(start)
    length = len(lst_move)    
    if length < 2:
        return 'Could not find possible path'
    else:
        for i in range(length-1):
            for j in range(i+1, length):
                path  = ''
                # Check if two char connect to the start or not
                if arr_tiles[lst_move[i][0], lst_move[i][1]] in dict_tile[lst_move[i][2]]:
                    path += lst_move[i][2]
                if arr_tiles[lst_move[j][0], lst_move[j][1]] in dict_tile[lst_move[j][2]]:
                    path += lst_move[j][2]
                
                if path in ('NS', 'SN'):                    
                    start_tile = '|'
                    next_tile = (lst_move[i][0], lst_move[i][1])
                elif path in ('WE', 'EW'):
                    start_tile = '-'
                    next_tile = (lst_move[i][0], lst_move[i][1])
                elif path in ('NE', 'EN'):
                    start_tile = 'L'
                    next_tile = (lst_move[i][0], lst_move[i][1])
                elif path in ('SE', 'ES'):
                    start_tile = 'F'
                    next_tile = (lst_move[i][0], lst_move[i][1])
                elif path in ('NW', 'WN'):
                    start_tile = 'J'
                    next_tile = (lst_move[i][0], lst_move[i][1])
                elif path in ('SW', 'WS'):
                    start_tile = '7'
                    next_tile = (lst_move[i][0], lst_move[i][1])
                    
    return start_tile, next_tile    
```

# Part 2
Part 1: By using DFS, you can easily solve this problem. After completing this part, you will have the path that creates a loop.

Part 2: There is a trick here; imagine the loop as a polygon, and the internal points are the integer points within that polygon. Using Pick's Theorem, we have:

S = I + B / 2 - 1

where:

S: the area of the polygon

I: the number of integer points inside the polygon

B: the number of integer points on the edge of the polygon

The area S can be easily calculated using the Shoelace formula, and B is the size of the path. Therefore, I can be calculated easily.

```python
# Using Pick theorem and Shoelace formula
S = 0
for i in range(len(lst_loop)-1):
    x1 = lst_loop[i][0]
    y1 = lst_loop[i][1]
    x2 = lst_loop[i+1][0]
    y2 = lst_loop[i+1][1]
    S  += (x2-x1)*(y2+y1)/2
print('Number of integer point ', S-(len(lst_loop)-1)/2 +1)
```
