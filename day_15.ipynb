```python
with open('data/day_15.txt', 'r') as f:
    input_data = f.read().split(',')
```

```python
def hash(str_input):
    middle_sum = 0
    for char in str_input:
        middle_sum += ord(char)
        middle_sum *= 17
        middle_sum = middle_sum%256
    return middle_sum
```

```python
total_sum = 0
for item in input_data:
    total_sum += hash(item)
print(total_sum)
```

```python
from collections import defaultdict


boxes = defaultdict(dict)
for cmd in input_data:
  if '-' in cmd:
    label = cmd[:-1]
    boxes[hash(label)].pop(label, None)
  else:
    label, i = cmd.split('=')
    boxes[hash(label)][label] = int(i)

print('part2: ', sum((i+1)*(j+1)*l for i in boxes for j,l in enumerate(boxes[i].values())))
```

```python
dict_box = {}
for item in input_data:
    if '=' in item:
        label, value = item.split('=')
        box = hash(label)
        if box not in dict_box.keys():
            dict_box[box] = [[label, value]]
        else:
            lst_box = dict_box[box]
            check = False
            for i, lst_value in enumerate(lst_box):
                if label == lst_value[0]:
                    lst_box[i] = [label, value]
                    check = True
                    break
            if check == False:
                lst_box.append([label, value])
    
    if '-' in item:
        label = item.split('-')[0]
        box = hash(label)
        
        if box in dict_box.keys():
            lst_box = dict_box[box]
            for i, lst_value in enumerate(lst_box):                
                if label == lst_value[0]:                    
                    lst_box.pop(i)
```

```python
total_sum = 0
for key, value in dict_box.items():
    middle_sum = 0
    for i, lst_value in enumerate(value):
        middle_sum += (i+1)*int(lst_value[1])
    total_sum += middle_sum*(key+1)
print(total_sum)
```
