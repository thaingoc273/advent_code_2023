## Part 1:
Using regrex to find all digit in the string and combine the first and the last digits

````python
def get_number(str_input):
    num = re.findall(r'\d', str_input) # Find all digit in the string
    if len(num)>0:
        return int(num[0])*10 + int(num[-1])
    else:
        return 0
````

````python
dict_number = {'one':1, 'two': 2, 'three':3, 'four': 4, 'five':5,
               'six':6, 'seven':7, 'eight':8, 'nine': 9}

def convert_to_number(str_input):
    if str_input in dict_number.keys():
        return dict_number[str_input]
    else:
        return int(str_input)

def get_number_2(str_input):
    num = re.findall(r'(?=(\d|one|two|three|four|five|six|seven|eight|nine))', str_input)
    if len(num)==0:
        return 0
    else:
        first_digit = convert_to_number(num[0])
        last_digit = convert_to_number(num[-1])
        return first_digit*10 + last_digit
````

````python
def sum_number(input_data):
    sum = 0
    for str in input_data:
        s = get_number(str)
        sum +=s
    return sum
````

````python
sum_number(input_data)
````
